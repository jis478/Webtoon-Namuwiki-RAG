## Import packages

```python
import os
import re
from datasets import load_dataset
os.environ["CUDA_VISIBLE_DEVICES"]="6"

```

## Dataset

```python
dataset = load_dataset("heegyu/namuwiki")

print(dataset["train"][0])
```

```python
for item in dataset["train"]:
    if item["title"] == "마음의소리":
        found_text = item["text"]
        break
```

## Data cleansing

```python

def extract_heading_content(text, heading):
    pattern_heading = re.compile(r'==\s*{}\s*==\s*(.*?)=='.format(heading), re.DOTALL)
    match_heading = pattern_heading.search(text)
    if match_heading:
        content = match_heading.group(1).strip()
        # Remove HTML tags
        content = re.sub(r'<[^>]*>', '', content)
        # Remove web addresses
        content = re.sub(r'\[?http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+\]?', '', content)
        # Remove "#"
        content = re.sub(r'#', '', content) 
        # Remove '파일:xxx.xxx'
        content = re.sub(r'파일:[^\s]+', '', content)
        # Remove "[[" and "]]"
        content = re.sub(r'\[\[|\]\]', '', content) 
        # Remove "'''" from the content
        content = re.sub(r"'''", '', content) # 강조문구 삭제
        # Remove texts enclosed between "[*" and "]"
        content = re.sub(r'\[\*.*?\]', '', content) # 주석 삭제
        # Remove "["
        content = re.sub(r'\[\|', '', content) # 중간 hyperlink 삭제
        # Remove texts beginning with ">"
        content = re.sub(r'>.*', '', content) # 초반 인용 link 삭제
        # Extract yyy from xxx|yyy pattern
        content = re.sub(r'\w+\|([^| ]+)', r'\1', content)
        content = re.sub(r'\w+\([^()]+\)\|([^| ]+)', r'\1', content)
        # Remove trailing empty lines
        content = content.replace('\n', '')
        return content
    else:
        return None

data_dict = {}
    
category = ["개요", "줄거리", "연재 현황", "완결", "등장인물", "인기", "평가"]

for cat in category:
    data_dict[cat] = extract_heading_content(found_text, cat)
```
